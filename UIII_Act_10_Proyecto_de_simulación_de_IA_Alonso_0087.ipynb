#### Cientifico y LosCientificos
##### puedes generar un programa en python que simule inteligencia artificial utilizando Una Lista de diccionarios, if. if else, elif, y while true ciclo for, referido a una lista de cientifico de dicionarrio loscientificos, tipo menu, que sea interactivo. No utilizar programación orientada a objetos, no funciones, no try, el ejemplo debe ser simple y atractivo para alumnos de preparatoria. al inicio la lista empieza vacía y también el diccionario vacio

```python
# --- SIMULADOR DE IA: GESTIÓN DE CIENTÍFICOS 'NEURAL-LAB 2026' ---

# Inicialización de la base de datos como una lista vacía [1]
loscientificos = []

# Ciclo infinito para mantener el menú interactivo activo [3]
while True:
    print("--- MENÚ SISTEMA IA DE LABORATORIO ---")
    print("1. Registrar nuevo científico")
    print("2. Listar investigadores registrados")
    print("3. Análisis de IA: Evaluación de Riesgo en Experimento")
    print("4. Salir")

    opcion = input("Seleccione una opción (1-4): ") # Recibe la entrada del usuario [4]
    print() # Espaciado para legibilidad

    # Opción 1: Crear y agregar un diccionario a la lista [2, 5]
    if opcion == "1":
        # Creamos un diccionario vacío para el nuevo registro [2]
        un_cientifico = {}

        # Solicitamos datos y los guardamos en claves del diccionario [5]
        un_cientifico["nombre"] = input("Nombre del científico: ")
        un_cientifico["especialidad"] = input("Especialidad (Física/Química/Bio): ")
        un_cientifico["años_exp"] = int(input("Años de experiencia: "))

        # Agregamos el diccionario completo a nuestra lista [1, 6]
        loscientificos.append(un_cientifico)
        print(f"SISTEMA: {un_cientifico['nombre']} ha sido registrado en la base de datos.")
        print()

    # Opción 2: Recorrer la lista de diccionarios con un ciclo for [7, 8]
    elif opcion == "2":
        print("--- LISTA DE INVESTIGADORES ACTIVOS ---")
        # El ciclo for recorre cada diccionario dentro de la lista [8, 9]
        for cientifico in loscientificos:
            # Accedemos a los valores usando las claves del diccionario [5]
            print(f"Científico: {cientifico['nombre']} | Especialidad: {cientifico['especialidad']}")

        # Operador relacional para verificar si no hay registros [10]
        if loscientificos == []:
            print("AVISO: No hay científicos registrados todavía.")
        print()

    # Opción 3: Simulación de IA utilizando lógica condicional y aritmética [11, 12]
    elif opcion == "3":
        print("--- MÓDULO DE ANÁLISIS DE RIESGO IA ---")
        temp = float(input("Ingrese la temperatura del reactor (°C): ")) # Entrada decimal [13]
        presion = float(input("Ingrese la presión detectada (atm): "))

        # La "IA" toma decisiones basadas en operadores relacionales y lógicos [7, 14]
        if temp > 100 or presion > 50: # Operador lógico 'or' para peligro [14]
            print("IA DIAGNÓSTICO: ¡ALERTA! El experimento es INESTABLE.")
        elif temp >= 20 and temp <= 25: # Operador lógico 'and' para rango óptimo [7]
            print("IA DIAGNÓSTICO: Condiciones AMBIENTALES perfectas.")
        else: # Caso por defecto si no se cumplen las anteriores [11]
            print("IA DIAGNÓSTICO: Condiciones estables. Proceda con precaución.")
        print()

    # Opción 4: Finalizar el programa [3]
    elif opcion == "4":
        print("Cerrando 'NEURAL-LAB 2026'. Sincronizando datos finales...")
        break # Rompe el ciclo while True para salir [3]

    # Manejo de opciones no válidas [11]
    else:
        print("ERROR: Opción no reconocida por la IA. Intente de nuevo.")
        print()
```
