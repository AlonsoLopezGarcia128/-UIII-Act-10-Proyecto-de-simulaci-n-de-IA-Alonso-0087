#### Cientifico y LosCientificos
##### puedes generar un programa en python que simule inteligencia artificial utilizando Una Lista de diccionarios, if. if else, elif, y while true ciclo for, referido a una lista de cientifico de dicionarrio loscientificos, tipo menu, que sea interactivo. No utilizar programación orientada a objetos, no funciones, no try, el ejemplo debe ser simple y atractivo para alumnos de preparatoria. al inicio la lista empieza vacía y también el diccionario vacio

```python
# --- SIMULADOR DE IA: GESTIÓN DE LABORATORIO 2026 ---

# Inicialización de las estructuras de datos solicitadas (Source 41, 55)
cientifico = []        # Lista vacía para nombres de investigadores (Source 41)
loscientificos = {}    # Diccionario vacío para especialidades (Source 55)

# Ciclo infinito para el menú interactivo (Source 517, 1052)
while True:
    print("--- MENÚ SISTEMA IA DE LABORATORIO ---")
    print("1. Registrar nombre en lista de asistencia (cientifico)")
    print("2. Registrar especialidad en base de datos (loscientificos)")
    print("3. Visualizar registros completos")
    print("4. Análisis de IA: Diagnóstico de Reactor")
    print("5. Salir")

    opcion = input("Seleccione una opción (1-5): ") # Recibe la opción del usuario
    print() # Espaciado solicitado para legibilidad

    # Opción 1: Agregar elementos a la LISTA (Source 41)
    if opcion == "1":
        nombre = input("Ingrese nombre del científico: ") # Lee el nombre
        cientifico.append(nombre) # El método append añade el dato al final de la lista (Source 41)
        print(f"IA NOTIFICACIÓN: {nombre} ha sido añadido a la lista de asistencia.")
        print()

    # Opción 2: Agregar elementos al DICCIONARIO (Source 56)
    elif opcion == "2":
        id_lab = input("Ingrese el ID del científico: ") # Lee la clave única
        rama = input("Ingrese su rama de estudio: ") # Lee el valor a guardar
        loscientificos[id_lab] = rama # Asigna el valor a la clave en el diccionario (Source 56)
        print(f"IA NOTIFICACIÓN: Registro de ID {id_lab} guardado con éxito.")
        print()

    # Opción 3: Recorrer estructuras con ciclos FOR (Source 60, 518)
    elif opcion == "3":
        print("--- LISTA DE ASISTENCIA (LISTA) ---")
        for c in cientifico: # Ciclo for para recorrer la lista (Source 518)
            print(f"- Investigador: {c}") # Muestra cada nombre
        if cientifico == []: # Operador relacional para verificar lista vacía (Source 451)
            print("AVISO: La lista de asistencia está vacía.")

        print("\n--- BASE DE ESPECIALIDADES (DICCIONARIO) ---")
        # El método items() permite obtener claves y valores simultáneamente (Source 58, 60)
        for matricula, especialidad in loscientificos.items():
            print(f"ID: {matricula} | Rama: {especialidad}") # Muestra los pares
        if loscientificos == {}: # Comprueba si el diccionario no tiene datos
            print("AVISO: El diccionario de especialidades está vacío.")
        print()

    # Opción 4: Lógica de IA con operadores aritméticos, lógicos y relacionales (Source 442, 451, 458)
    elif opcion == "4":
        print("--- MÓDULO DE ANÁLISIS TÉRMICO IA ---")
        temp = float(input("Ingrese temperatura del reactor (°C): ")) # Entrada decimal
        presion = float(input("Ingrese presión del sistema (psi): ")) # Entrada numérica

        # La IA toma decisiones basadas en operadores relacionales (>, ==) y lógicos (and, or) (Source 451, 458)
        if temp > 120 and presion > 80: # Ambas condiciones deben ser verdaderas (Source 457)
            print("¡ALERTA CRÍTICA!: Peligro de explosión detectado.")
        elif temp < 0 or presion == 0: # Una de las condiciones debe ser verdadera (Source 458)
            print("IA REPORTE: El sistema se encuentra inactivo o congelado.")
        else: # Caso por defecto
            print("IA REPORTE: Funcionamiento dentro de rangos normales.")

        # Ejemplo de cálculo con operadores aritméticos (+, *) (Source 442)
        ajuste = (temp * 1.05) + 5 # Realiza un cálculo de compensación rápida
        print(f"Sugerencia de ajuste de flujo: {ajuste:.2f}")
        print()

    # Opción 5: Finalización del programa (Source 539, 1052)
    elif opcion == "5":
        print("Desactivando Sistema de IA. Sincronizando datos finales...")
        break # Rompe el ciclo while True para cerrar el programa (Source 1052)

    # Control de errores para opciones no listadas
    else:
        print("ERROR: Comando no reconocido por la interfaz. Reintente.")
```
